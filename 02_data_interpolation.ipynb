```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from hampel import hampel
import pickle

plt.style.use('ggplot')
myColor  = ("#E3072A","#0A8974","#5FAECC","#E49C23","#E27520") # Darjeeling Color map
```

We load the data file that we generated with "01_clean_data.ipynb"

```python
# Load results

path = 'Results/pkl_files/'

with open(path + 'cleaned_data.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
keys = data.keys()

int_data = {}

fig, ax = plt.subplots(nrows= len(keys),ncols=1,figsize=(14.0,5.0), sharex = True) # ,sharey=True

for i, key in enumerate(keys):

    df = data[key]
    df['date'] = pd.to_datetime(df['date'])
    df = df.set_index('date')
    # Resampling
    df = df.resample('60T').mean()
    # Interpolation
    df = df.interpolate(method='polynomial', order=3)
    
    # plot
    ax[i].plot(df.index,df['head[m]'], color='k',label = key + ' interpolated')
    ax[i].plot(data[key]['date'][1::50], data[key]['head[m]'][1::50],'.', label = key, color=myColor[i])
    ax[i].set_ylabel('Head [m]')  
    ax[i].legend()
    
    # save in dic
    int_data[key] = df
    
ax[i].set_xlabel('Time [months]')  
```

```python
# Save data

path = 'Results/pkl_files/'

with open(path + 'interpolated_data.pkl', 'wb') as f:
    pickle.dump(int_data, f)
```
